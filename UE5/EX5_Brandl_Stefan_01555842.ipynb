<h3 style="color:rgb(208,90,80)">Answer 5.1:</h3>

- **The VC-dimension of the hypothesis class of intervals is $2$**

For $0$ and $1$ points this holds trivially. 

We classify a point $x$ as $1$ $\iff$ $\exists a, b | a < x < b$ with $x \in \mathbb{R}$. 

We classify a point as $0$ if it is outside this interval.

For $x_1 < x_2$ we get the following intervalls 



| labeling | $x_1$ | $x_2$ | possible interval   |
|----------|-------|-------|---------------------|
| 1        | 0     | 0     | $(x_1, x_2)$        |
| 2        | 0     | 1     | $(x_1, \infty)$     |
| 3        | 1     | 0     | $(-\infty, x_2)$    |
| 4        | 1     | 1     | $(-\infty, \infty)$ |

For three points however, this is no longer possible

| labeling | $x_1$ | $x_2$ | $x_3$ | possible interval   |
|----------|-------|-------|-------|---------------------|
| 1        | 0     | 0     | 0     | $(x_3, \infty)$     | 
| 2        | 0     | 0     | 1     | $(x_2, \infty)$     | 
| 3        | 0     | 1     | 0     | $(x_1, x_3)$        | 
| 4        | 0     | 1     | 1     | $(x_1, \infty)$     | 
| 5        | 1     | 0     | 0     | $(-\infty, x_2)$    | 
| 6        | 1     | 0     | 1     | impossible          | 
| 7        | 1     | 1     | 0     | $(-\infty, x_3)$    | 
| 8        | 1     | 1     | 1     | $(-\infty, \infty)$ | 

Thus we see that the $VC_{dim}$ of intervals is $2$.

- **The VC-dimension of the hypothesis class of unions of $k$  intervals is 2$k$.**

Assuming that all intervalls are well ordered and do not overlap we could (in the worst case) end up with alternating $10$ labels (in regex syntax: $(1^+0^+)^k$ *one or more $1$ followed by one or more $0$, repeated $k$ times*).

Then we can iteratively split one label series of either $1$s or $0$s from the union. 
Since there are $2$ series per set over $k$ sets in total, at worst there are $2k$ splits necessary. 

Thus the $VC_{dim}$ of $k$ unions of intervals is $2k$.

```python
# Some code that should help you. Nothing to do here
import sklearn as sk
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
# Create empty list that should store feature values.
x = []
# Create empty list that should store labels.
y = []
# Randomly create one point on unit circle with label y=1, store features and labels in the corresponding lists.
length_1 = np.random.uniform(0, 1)
angle_1 = np.pi * np.random.uniform(0, 2)
x.append(
    [np.sqrt(length_1) * np.cos(angle_1),
     np.sqrt(length_1) * np.sin(angle_1)])
y.append(1)

#Randomly create another point on unit circle with label y=-1, store features and labels it in the corresponding lists.
length_2 = np.random.uniform(0, 1)
angle_2 = np.pi * np.random.uniform(0, 2)
x.append(
    [np.sqrt(length_2) * np.cos(angle_2),
     np.sqrt(length_2) * np.sin(angle_2)])
y.append(-1)
print(x, y)

#Code for linear SVC containing only the previously created points as data, also outputs corresponding margin.
clf = svm.SVC(kernel='linear', C=1)
clf.fit(x, y)
margin = 1 / np.linalg.norm(clf.coef_)
print(margin)
```

<h3 style="color:rgb(208,90,80)">Answer 5.3, 5.4 and 5.5:</h3>

```python
def k_circle_points(k):
    data = []
    for i in range(k):
        angle = np.pi * np.random.uniform(high=2)
        data.append((np.cos(angle), np.sin(angle)))
    data = np.asarray(data)
    return data, (data[:, 1] > 0).astype(int)

X, y = k_circle_points(200)
plt.figure(figsize=(10, 10))
plt.scatter(X[:, 0], X[:, 1], c=[['blue', 'red'][y_i] for y_i in y])
plt.xlabel('x')
plt.ylabel('y')
plt.title('200 sample points')
plt.show()
```

```python
def compute_svm(k):
    if k <= 10:
        raise ValueError(f'k needs to be >= 10; but was: {k}')
        
    clf = svm.SVC(kernel='linear')
    margins = []
    for j in range(10, k + 1):
        X, y = k_circle_points(j)
        clf.fit(X, y)
        margins.append(1 / np.linalg.norm(clf.coef_))
    
    return margins

marg = np.asarray(compute_svm(1000))
print(f'First 5: {marg[:5]}')
print('...')
print(f'Last 5: {marg[-5:]}')
```

```python
bounds = 1 / marg**2

_, (ax_m, ax_b) = plt.subplots(1, 2, figsize=(16, 9))

ax_m.set_title('Margins')
ax_m.plot(marg, color='orange', alpha=0.7, linewidth=0.5)
ax_m.set_ylabel('gamma')
ax_m.set_xlabel('k')
ax_m.set_xticks(ticks=range(10, 1001, 90))

ax_b.set_title('VC-dimension')
ax_b.plot(bounds, color='orange', alpha=0.7, linewidth=0.5)
ax_b.set_ylabel('bounds')
ax_b.set_xlabel('k')
ax_b.set_xticks(ticks=range(10, 1001, 90))

plt.show()
```

It should be clearly visible from the plot, that, up to small fluctuations, the value of the margin decreases for a larger number of data points. This should come as no surprise, since the model gets increasingly complex for a larger number of data, and therefore a clearer separation is harder, which is indicated by smaller margins. Also the increase of the VC-dimension should come as a reasonable fact, since VC-dimension provides an indicator for model complexity and it indeed seems to be a reasonable concept that can provide a theoretical basis for measuring the quality of widle applied classifier algorithms.
