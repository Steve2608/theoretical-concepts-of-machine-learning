<h3 style="color:rgb(0,120,170)">Question 1.1:</h3>
We are given a random variable $X$ with pdf
\begin{align*}
f(x ; p) &= p^x(1-p)^{1-x} \ , \\
\end{align*}
where $0 \leq p \leq 1$ is the parameter and the support is $x \in \{0,1\}$. 

* Do you know the name of the distribution to which the pdf belongs?
* Calculate the Fisher information!  
Assume this distribution fullfils all the required regularity conditions, i.e. you can use the following formula:
\begin{align*}
I_F(p) &= -\mathbb{E}_X \big[ \frac{d^2}{dp^2} \log f(x ; p) \big] \ .
\end{align*}

<h3 style="color:rgb(208,90,80)">Answer 1.1:</h3>

**Do you know the name of the distribution to which the pdf belongs?**

It is a special case of the general binomial distribution - the Bernoulli Distribution: 
$$
\binom{n}{k} p^k (1 - p)^{n - k} ~~~~ \text{with $n=1$ and $k \in \{0, 1\}$} \\
1 p^k (1 - p)^{1 - k} ~~~~ \text{renaming $k$ to $x$} \\
p^x (1 - p)^{1 - x}
$$

**Calculate the Fisher information!**
\begin{align*}
I_F(p) &= -\mathbb{E}_X \big[\frac{\partial ^2}{\partial p^2} \log f(x;p) \big] 
\end{align*}

Firstly, we calculate the first derivative:
$$
\frac{\partial}{\partial p}[x \ln(p) + (1-x)\ln(1-p)] = \frac{p - x}{(p - 1) p}
$$

the second derivative:
$$
\frac{\partial}{\partial p}[\frac{p - x}{(p - 1) p}] = -\frac{p^2 - 2xp + x}{(p - 1)^2 p^2}
$$

Negative expectation:
\begin{align*}
I_F(p) &= -\mathbb{E}_X \big[-\frac{p^2 - 2xp + x}{(p - 1)^2 p^2}\big] \\
&= \mathbb{E}_X \big[ \frac{p^2 - 2xp + x}{(p - 1)^2 p^2} \big] \\
&= \frac{-p^2 + p}{(p - 1)^2 p^2} = \frac{-p (p - 1)}{(p - 1)^2 p^2}\\
&= -\frac{1}{p(p - 1)} = \frac{1}{p(1 - p)}
\end{align*}

Suppose we draw $n$ i.i.d samples  $\{x_1, \ldots, x_n \}$ from the above random variable $X$. Then the Likelihood is
\begin{align*}
f(\{x_1, \ldots, x_n \}; p) &= \prod^n_{i=1} f(x_i; p) = \prod^n_{i=1} p^{x_i}(1-p)^{1-{x_i}} \ .\\
\end{align*}

* Calculate the Fisher Information, that is, calculate 
\begin{align*}
I_F(p) &= - \mathbb{E}_X \big[ \frac{\partial^2}{\partial p^2} \log f(\{x_1, \ldots, x_n \}; p) \big] \ . 
\end{align*}

We first take the first derivative:
\begin{align*}
\frac{\partial}{\partial p} \log \prod_i^n p^{x_i} (1 - p)^{1 - x_i} &= \frac{\partial}{\partial p} \sum_i^n \log p^{x_i} (1 - p)^{1 - x_i} \\
&= \frac{\partial}{\partial p} \sum_i^n \log(p^{x_i}) + \log((1 - p)^{1 - x_i}) \\
&= \frac{\partial}{\partial p} \sum_i^n x_i \log(p) + (1 - x_i) \log(1 - p)) \\
&= \frac{\partial}{\partial p} \sum_i^n x_i \log(p) + (1 - x_i) \log(1 - p)) \text{, from earlier we know} \\
&= \sum_i^n \frac{p - x_i}{(p - 1) p}
\end{align*}

the second derivative:
$$
\frac{\partial}{\partial p}[\sum_i^n \frac{p - x_i}{(p - 1) p}] = \sum_i^n -\frac{p^2 - 2x_ip + x_i}{(p - 1)^2 p^2} = -\sum_i^n \frac{p^2 - 2x_ip + x_i}{(p - 1)^2 p^2}
$$

Analogously to above:
\begin{align*}
I_F(p) &= -\mathbb{E}_X \big[-\sum_i^n \frac{p^2 - 2x_ip + x_i}{(p - 1)^2 p^2}\big] \\
&= \mathbb{E}_X \big[ \sum_i^n \frac{p^2 - 2x_ip + x_i}{(p - 1)^2 p^2}\big] \\
&= \sum_i^n \frac{-p^2 + p}{(p - 1)^2 p^2} = \sum_i^n \frac{-p (p - 1)}{(p - 1)^2 p^2}\\
&= \sum_i^n -\frac{1}{p(p - 1)} = -n \frac{1}{p(p - 1)} = -\frac{n}{p(p - 1)} = \frac{n}{p(1 - p)}
\end{align*}

# Exercise 2

We are going to take a look at two different estimators for the population variance.
Write a function that samples $n$ numbers from a Gaussian distribution with mean $\mu=5$ and variance $\sigma^2=4$. 
Then, calculate the variance of your sample using the following two 
estimators ($\bar{x}$ denotes the sample mean of the data):
\begin{align*}
   (i) \quad \hat{\sigma}^2_A = \frac{1}{n-1}\sum_i (\bar{x} - x_i)^2
\end{align*}
and
\begin{align*}
   (ii) \quad \quad \ \hat{\sigma}^2_B = \frac{1}{n}\sum_i (\bar{x} - x_i)^2
\end{align*}

Do this 10\,000 times, drawing $n$ new samples each time, and keep track of the mean and variance of both estimators
(Don't get confused: we want you to calculate the variance of the estimates of the variance :) ). 
Do this for $n$ = 10 and $n$ = 1000.

* On average, which estimator is closer to the true value? And which one has less variance?
* What is the Cramer-Rao Lower Bound for this estimation task? You will probably have
 observed that one of the estimators has even less variance than the CRLB predicts. How can that be?
 
Hint: the Fisher Information Matrix for a sample of size $n$ of a Gaussian distribution is:
\begin{align*}
I_F = \left(
    \begin{array}{cc}
    \frac{n}{\sigma^2} & 0 \\
    0 & \frac{n}{2 \sigma^4} \end{array} \right) \ .
\end{align*}

```python
import numpy as np
from numpy.random import normal
import matplotlib.pyplot as plt

def draw(n, loc=5, scale=2):
    return normal(loc, scale, n)

def est1(data):
    return np.sum((data - data.mean())**2) / (len(data) - 1) 

def est2(data): 
    return np.sum((data - data.mean())**2) / len(data) 

def plot_stats(d1, d2, l1, l2, e, l3, title):
    plt.figure(figsize=(16, 7))
    plt.plot(d1, linewidth=1)
    plt.plot(d2, linewidth=1)
    plt.hlines(e, xmin=0, xmax=len(d1), linewidth=1, linestyle='--')
    plt.legend([l1, l2, l3])
    plt.title(title)
    plt.show()
```

```python
for n in [10, 1000]:
    m1, m2 = [], []
    var1, var2 = [], []
    varvar1, varvar2 = [], []
    for i in range(10000):
        data = draw(n)
        
        var1 += [est1(data)]
        var2 += [est2(data)]
        m1 += [np.mean(var1)]
        m2 += [np.mean(var2)]
        varvar1 += [np.var(var1)]
        varvar2 += [np.var(var2)]

    plot_stats(m1, m2, 'Mean1', 'Mean2', 4, 'CRLB', f'Mean of Variances({n})')
    plot_stats(varvar1, varvar2, 'Var1', 'Var2', 32/n, 'CRLB', f'Variance of Variances({n})')
```

<h3 style="color:rgb(208,90,80)">Answer 2.2:</h3>

**On average, which estimator is closer to the true value? And which one has less variance?**

The first estimator $\hat\sigma_1$ is closer to the real value (given by the CRLB). Since the variance is computed identically (only the factors $\frac{1}{n-1}$ and $\frac{1}{n}$ are different) estimator $\hat\sigma_2$ has a lower variance (bigger divisor). 

**What is the Cramer-Rao Lower Bound for this estimation task? You will probably have observed that one of the estimators has even less variance than the CRLB predicts. How can that be?**

The CRLB is the inverse of the fisher information matrix: 

mean: $\frac{1}{\frac{n}{\sigma^2}} = \frac{\sigma^2}{n} = \frac{4}{n}$

variance: $\frac{1}{\frac{n}{2 \sigma^4}} = \frac{2 \sigma^4}{n} = \frac{32}{n}$

The second estimator $\hat\sigma_2$ predicts a lower value than the CRLB predicts, because it is biased. 

The variance can be defined as follows: 

\begin{align*}
\sum_i^n (x_i - \overline{x})^2 &= \sum_i^n x_i^2 - 2x_i\overline{x} - \overline{x}^2 \\
&= \sum_i^n x_i^2 - n\overline{x}^2
\end{align*}

\begin{align*}
\mathbb{E}(\sigma^2) &= \mathbb{E}(\sum_i^n \frac{(x_i - \overline{x})^2}{n - 1}) \\
&= \frac{1}{n - 1} \mathbb{E}(\sum_i^n x_i^2 - n\overline{x}^2) \\
&= \frac{1}{n - 1} \sum_i^n (\text{Var}(x_i) - n\text{Var}(\overline{x}))
\end{align*}

We also know:
\begin{align*}
\text{Var}(x_i) = \sigma^2 & &  \text{Var}(\overline{x}) = \frac{\sigma^2}{n}
\end{align*}

Thus:
\begin{align*}
\mathbb{E}(\sigma^2) &= \frac{1}{n - 1} \sum_i^n (\text{Var}(x_i) - n\text{Var}(\overline{x})) \\
&= \frac{1}{n - 1} \sum_i^n (\sigma^2) - n\frac{\sigma^2}{n} \\
&= \frac{1}{n - 1} (n\sigma^2 - \sigma^2) = \frac{1}{n - 1} (n-1)\sigma^2 = \sigma^2
\end{align*}
So we can see that the expectation of the variance equals the true variance. 

For $\frac{1}{n}$ however:
\begin{align*}
\mathbb{E}(\sigma^2) &= \frac{1}{n} (n-1)\sigma^2 = \frac{n-1}{n} \sigma^2 
\end{align*}
We can see that this, however, is not the real variance - thus the estimator $\hat\sigma_2$ is biased.
